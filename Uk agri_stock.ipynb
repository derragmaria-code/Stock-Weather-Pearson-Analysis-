```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# 1. Pre-war Data (2019-2021)
# Note: 2021 was a record harvest year for Ukraine due to favorable weather.
data = {
    'Year': [2019, 2020, 2021],
    # Avg Rainfall (mm) for the critical growing window (April-June)
    # 2020 saw a significant spring drought in the Steppe region.
    'Spring_Rainfall_mm': [155.0, 110.5, 195.2], 
    # Agricultural GDP contribution in Billion UAH (Estimated pre-war values)
    'Agri_GDP_Billion_UAH': [450.2, 410.8, 580.5],
    # Representative Stock Price (e.g., Kernel Holding - KEL) in PLN/Share
    'Kernel_Stock_Price': [48.50, 42.10, 61.20]
}

df = pd.DataFrame(data)

# 2. Statistical Calculation (Pearson Correlation)
# We test the link between rain and the specific stock price
corr_coef, p_value = pearsonr(df['Spring_Rainfall_mm'], df['Kernel_Stock_Price'])
r_squared = corr_coef**2

# 3. Output the Findings
print(f"--- Pre-War Ukraine Analysis (2019-2021) ---")
print(f"Pearson Correlation (r): {corr_coef:.4f}")
print(f"Coefficient of Determination (RÂ²): {r_squared:.4f}")
print(f"Interpretation: {r_squared*100:.1f}% of stock variance associated with Spring Rain.")

# 4. Visualization
fig, ax1 = plt.subplots(figsize=(10, 6))

# Primary Axis: Rainfall
color = 'tab:blue'
ax1.set_xlabel('Year')
ax1.set_ylabel('Spring Rainfall (mm)', color=color)
ax1.bar(df['Year'], df['Spring_Rainfall_mm'], color=color, alpha=0.3, label='Rainfall')
ax1.tick_params(axis='y', labelcolor=color)

# Secondary Axis: Stock Price
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Stock Price (Kernel Holding)', color=color)
ax2.plot(df['Year'], df['Kernel_Stock_Price'], color=color, marker='o', linewidth=2, label='Stock Price')
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Ukraine Pre-War: Spring Rainfall vs. Agricultural Stock Performance')
plt.xticks([2019, 2020, 2021])
fig.tight_layout()
plt.show()
```
