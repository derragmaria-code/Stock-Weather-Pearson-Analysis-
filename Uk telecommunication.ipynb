```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# 1. Pre-war Telecom & Weather Data (2019-2021)
# Note: Telecom is less seasonal than agriculture, so we use Annual Average Rainfall.
data = {
    'Year': [2019, 2020, 2021],
    # Annual Avg Rainfall in Ukraine (mm)
    # 2020 was historically dry; 2021 was wetter.
    'Annual_Rainfall_mm': [549, 450, 605], 
    # Combined Revenue of Big 3 (Kyivstar, Vodafone, Lifecell) in Billion UAH
    # Pre-war telecom showed steady growth due to 4G expansion.
    'Telecom_Revenue_Billion_UAH': [45.8, 49.2, 53.5],
    # Proxy Stock Index (e.g., VEON - Parent of Kyivstar) in USD
    'VEON_Stock_Price': [2.45, 1.85, 1.70]
}

df = pd.DataFrame(data)

# 2. Correlation: Does more rain = lower stock/revenue due to 'signal fade'?
corr_coef, _ = pearsonr(df['Annual_Rainfall_mm'], df['VEON_Stock_Price'])

# 3. Visualization
fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.set_xlabel('Year')
ax1.set_ylabel('Annual Rainfall (mm)', color='tab:blue')
ax1.plot(df['Year'], df['Annual_Rainfall_mm'], color='tab:blue', marker='s', label='Rainfall')
ax1.tick_params(axis='y', labelcolor='tab:blue')

ax2 = ax1.twinx()
ax2.set_ylabel('Telecom Stock Proxy (USD)', color='tab:red')
ax2.plot(df['Year'], df['VEON_Stock_Price'], color='tab:red', marker='o', linestyle='--', label='Stock Price')
ax2.tick_params(axis='y', labelcolor='tab:red')

plt.title('Ukraine Telecom: Weather vs. Digital Pulse (2019-2021)')
plt.xticks([2019, 2020, 2021])
print(f"Pearson Correlation (Rain vs. Telecom Stock): {corr_coef:.4f}")
plt.show()
```
